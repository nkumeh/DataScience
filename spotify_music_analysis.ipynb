```python
# import the json file

import pandas as pd

# Replace 'YourFileName.json' with the name of your uploaded file
my_spot_data = pd.read_json('/content/drive/MyDrive/MusicAnalysis/StreamingHistory0.json')

# Display the first few rows to check the data
my_spot_data.head()
```

```python
# data cleaning, transformation, converting datatime to date

# Convert the string to datetime format
# my_spot_data['endTime'] = pd.to_datetime(my_spot_data['endTime']).dt.date

# Extracting date, year, month, and day:
my_spot_data['endTime'] = pd.to_datetime(my_spot_data['endTime'])
my_spot_data['year'] = my_spot_data['endTime'].dt.year
my_spot_data['month'] = my_spot_data['endTime'].dt.month
my_spot_data['day'] = my_spot_data['endTime'].dt.day
```

```python
# cleaned dataframe
my_spot_data.head()
```

```python
# top tracks

# Returns a list containing counts of unique rows in the DataFrame.
frequency_top_tracks = my_spot_data['trackName'].value_counts().head(10)

relative_freq_top_tracks = my_spot_data['trackName'].value_counts(normalize=True).head(10)

print('Top 10 Tracks: frequency vs relative frequency:')
# print(top_tracks)
frequency_top_tracks, relative_freq_top_tracks
```

```python
# top artists

top_artists = my_spot_data['artistName'].value_counts().head(10)
print("\nTop 10 Artists:")
print(top_artists)
```

```python
# process the data into csv

my_spot_data.to_csv('processed_spotify_data.csv', index=False)
```
